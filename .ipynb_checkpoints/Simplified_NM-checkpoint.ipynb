```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint ,  solve_ivp
import matplotlib.ticker as plticker
```

```python
g_y=1.24e6   ##Hz
omega_0=4.45e15 ##Hz
del_omega=1e8
omega_1=omega_0-del_omega ##Hz
ohm=1e6         ##Hz  ## mechanical frequency
g_z0=0
g_z1=(20*ohm)**0.5
delta_0=100*ohm
N0=1e1                ## photon number at equillibrium

g_y/=ohm
g_z1/=ohm
g_z0/=ohm
omega_1/=ohm
omega_0/=ohm
del_omega/=ohm
```

```python
#z0 -> initial condition
#tf -> final time
#param_list -> list of paramters of system
#N_step -> number of time steps
#method -> method used to simulate differential equations


def psi_evol_app(z0,tf,param_list,N_step=300,method="DOP853"):
    
    def eig_to_real(zz):
        ## function to convert eigenmodes to physical variables
        temp=np.array(zz,dtype=complex)
        N_step=len(zz[0])
        yy=[]
        for i in range(N_step):
            yy.append(np.einsum("ij,j",U,zz[:4,i]))
        temp[:4]=np.array(yy).T
        return temp 
    
    def energy(yy):
        ## function to calculate energy
        q_y,p_y,A1_p,A1_m,A0_p,A0_m,q_z,p_z=yy[:8]        
        sigma_m=2*(A1_p*A0_m-A1_m*(A0_p+1))
        E=N0*(2*A0_p+A0_p**2+A0_m**2)*(del_omega-g_z1*q_z)+g_z1*q_z*N0+(q_y**2+p_y**2+q_z**2+p_z**2)/4+q_y*g_y*N0*sigma_m/2
        return E
    #t -> time (dont't remove)
    #z -> system variables (don't remove)
    #others -> parameters, can be changed freely
    def psi_ODE(t,z,g_z1,g_y,del_omega,N0):
        ## set of ODE for eigenmodes
        D=((1-del_omega)**2-8*g_y**2*N0*del_omega)**0.5
        lam=((del_omega**2+D)/2)**0.5
        nu=((del_omega**2-D)/2)**0.5
        psi_1,psi_1_d,psi_3,psi_3_d,A0_p,A0_m,q_z,p_z=z
        psi_1_dot = (-1j*lam+1j*g_z1*q_z*nu**2/(2*lam**2)+3*(nu**2/(2*lam))*(A0_m+1j*A0_p))*psi_1-(nu**2/(2*lam))*(A0_m-3*1j*A0_p)*psi_1_d+1j*g_z1*q_z*(psi_1_d+(psi_3+psi_3_d)/2)+lam*(A0_m+1j*A0_p)*(psi_3+psi_3_d)/2
        psi_1_d_dot= (1j*lam-1j*g_z1*q_z*nu**2/(2*lam**2)+3*(nu**2/(2*lam))*(A0_m-1j*A0_p))*psi_1_d-(nu**2/(2*lam))*(A0_m+3*1j*A0_p)*psi_1-1j*g_z1*q_z*(psi_1+(psi_3+psi_3_d)/2)+lam*(A0_m-1j*A0_p)*(psi_3+psi_3_d)/2

        psi_3_dot  =-1j*nu*(psi_3+(3*A0_p/2+g_z1*q_z/(2*lam))*(psi_3+psi_3_d)+(A0_p+g_z1*q_z/(2*lam))*(psi_1+psi_1_d)-1j*A0_m*(psi_1-psi_1_d)-1j*nu/(2*lam)*A0_m*(3*psi_3-psi_3_d))
        psi_3_d_dot= 1j*nu*(psi_3_d+(3*A0_p/2+g_z1*q_z/(2*lam))*(psi_3+psi_3_d)+(A0_p+g_z1*q_z/(2*lam))*(psi_1+psi_1_d)-1j*A0_m*(psi_1-psi_1_d)+1j*nu/(2*lam)*A0_m*(3*psi_3_d-psi_3))
        
        A0_m_dot=-(nu**2/(2*lam))*(psi_1+psi_1_d)**2-lam*(psi_3+psi_3_d)*(psi_1+psi_1_d+psi_3+psi_3_d)/2
        A0_p_dot=-1j*(psi_3+psi_3_d)*(lam*(psi_1-psi_1_d)+nu*(psi_3-psi_3_d))/2-1j*(nu**2/(2*lam))*(psi_1**2-psi_1_d**2)
        q_z_dot=p_z
        p_z_dot=-q_z-g_z1*N0*(-lam*(psi_1-psi_1_d)**2+(psi_3+psi_3_dot)**2+2*(psi_1+psi_1_d)*(psi_3+psi_3_d)-2*nu*(psi_1-psi_1_d)*(psi_3-psi_3_d))
#         N_dot=2*A0_p_dot+2*A0_p*A0_p_dot+2*A0_m*A0_m_dot+A1_p*A1_p_dot+A1_m*A1_m_dot

        x_list=[psi_1_dot,psi_1_d_dot,psi_3_dot,psi_3_d_dot,A0_p_dot,A0_m_dot,q_z_dot,p_z_dot]
        return x_list

    ## constructing eigen-matrices to verify if everything is in order
    g_z1,g_y,del_omega,N0=param_list
    D=((1-del_omega**2)**2-8*g_y**2*N0*del_omega)**0.5
    lam=((1+del_omega**2+D)/2)**0.5
    nu=((1+del_omega**2-D)/2)**0.5  
    print("$\\nu=$",nu,'$\\lambda=$',lam)
    lp=(1-del_omega**2+D)/(2*del_omega*g_y)
    lm=(1-del_omega**2-D)/(2*del_omega*g_y)
    ## evolution matrix in linear case
    U=np.array([
                [lp,lp,lm,lm],
                [-1j*lp*lam,1j*lp*lam,-1j*lm*nu,1j*lm*nu],
                [-1j*lam/del_omega,1j*lam/del_omega,-1j*nu/del_omega,1j*nu/del_omega],
                [1,1,1,1]
    ])
    
    Eig=np.diag([-1j*lam,1j*lam,-1j*nu,1j*nu])
    t = np.linspace(0,tf,N_step)
    z0[:4]*=del_omega*g_y/(2*D)
    #######MOST IMPORTANT##########
    #solve_ivp->sci_py solver
    #psi_ODE -> set of linear differential equations
    #[0,tf] -> time interval
    #z0 -> initial conditions
    #args -> parameters for the function psi_ODE
    #dense_output -> not of our concern
    # rtol , atol -> error tolerance
    # t_eval -> times to be evaluated at
    # method -> method used to simulate differential equations
    xx=solve_ivp(psi_ODE, [0,tf],z0,args=param_list,dense_output=False,rtol = 1e-8, atol = 1e-12, t_eval=t,method=method)
    ###############################
    zz=xx.y
    yy=np.real(eig_to_real(zz))
    H=energy(yy)
    sol_real=[]
    sol_eig=[]
    sol_real.extend( (yy[6],yy[7],yy[0],yy[1],yy[4],yy[5],yy[2],yy[3]) )
    sol_eig.extend((zz[1],zz[3],zz[5],zz[7]))
    #sol_real -> solution in terms of actual variables
    #sol_eig -> solution in terms of eigenmodes
    #t -> time interval
    return sol_real,sol_eig,t
```

```python
# exactly simulations of eigenmodes psi
def psi_evol(z0,tf,param_list,N_step=300,method="DOP853"):
    
    def eig_to_real(zz):
        temp=np.array(zz,dtype=complex)
        N_step=len(zz[0])
        yy=[]
        for i in range(N_step):
            yy.append(np.einsum("ij,j",U,zz[:4,i]))
        temp[:4]=np.array(yy).T
        return temp 
    
    def energy(yy):
        q_y,p_y,A1_p,A1_m,A0_p,A0_m,q_z,p_z=yy[:8]        
        sigma_m=2*(A1_p*A0_m-A1_m*(A0_p+1))
        E=N0*(2*A0_p+A0_p**2+A0_m**2)*(del_omega-g_z1*q_z)+g_z1*q_z*N0+(q_y**2+p_y**2+q_z**2+p_z**2)/4+q_y*g_y*N0*sigma_m/2
        return E


    def psi_ODE_v1(t,z,g_z1,g_y,del_omega,N0):
        
        psi_1,psi_2,psi_3,psi_4,A0_p,A0_m,q_z,p_z,N=z
        psi=np.array([psi_1,psi_2,psi_3,psi_4])
        w=(U@psi)
        q_y,p_y,A1_p,A1_m=w
        nl=np.array([0,-2*g_y*N0*(A1_p*A0_m-A1_m*A0_p),+g_z1*q_z*A1_m-g_y*q_y*A0_p,-g_z1*q_z*A1_p-g_y*q_y*A0_m],dtype=complex)
        psi_dot=Eig@psi+U_inv@nl

        q_z_dot=p_z
        p_z_dot=-q_z+2*g_z1*N0*(2*A0_p+A0_p**2+A0_m**2)

        A0_p_dot=g_y*q_y*A1_p/2
        A0_m_dot=g_y*q_y*A1_m/2

        psi_1_dot,psi_2_dot,psi_3_dot,psi_4_dot=psi_dot
        w_dot=(U@psi_dot)
        q_y_dot,p_y_dot,A1_p_dot,A1_m_dot=w_dot
        N_dot=2*A0_p_dot+2*A0_p*A0_p_dot+2*A0_m*A0_m_dot+A1_p*A1_p_dot+A1_m*A1_m_dot

        x_list=[psi_1_dot,psi_2_dot,psi_3_dot,psi_4_dot,A0_p_dot,A0_m_dot,q_z_dot,p_z_dot,N_dot]
        return x_list

    g_z1,g_y,del_omega,N0=param_list
    
    D=((1-del_omega**2)**2-8*g_y**2*N0*del_omega)**0.5
    lam=((1+del_omega**2+D)/2)**0.5
    nu=((1+del_omega**2-D)/2)**0.5
    print("$\\nu=$",nu,'$\\lambda=$',lam)

    lp=(1-del_omega**2+D)/(2*del_omega*g_y)
    lm=(1-del_omega**2-D)/(2*del_omega*g_y)

    U_inv=(del_omega*g_y/(2*D))*np.array([
                [1,1,-1,-1],
                [1j/lam,-1j/lam,-1j/nu,1j/nu],
                [-1j*del_omega*lm/lam,1j*del_omega*lm/lam,1j*del_omega*lp/nu,-1j*del_omega*lp/nu],
                [-lm,-lm,lp,lp]
        ]).T

    U=np.array([
                [lp,lp,lm,lm],
                [-1j*lp*lam,1j*lp*lam,-1j*lm*nu,1j*lm*nu],
                [-1j*lam/del_omega,1j*lam/del_omega,-1j*nu/del_omega,1j*nu/del_omega],
                [1,1,1,1]
    ])
    
    Eig=np.diag([-1j*lam,1j*lam,-1j*nu,1j*nu])
    t = np.linspace(0,tf,N_step)
    z0[:4]*=del_omega*g_y/(2*D)
    xx=solve_ivp(psi_ODE_v1, [0,tf],z0,args=param_list,dense_output=False,rtol = 1e-8, atol = 1e-12, t_eval=t,method=method)
    zz=xx.y
    yy=np.real(eig_to_real(zz))
    H=energy(yy)
    Num=yy[8]
    sol_real=[]
    sol_eig=[]
    sol_real.extend( (yy[6],yy[7],yy[0],yy[1],yy[4],yy[5],yy[2],yy[3]) )
    sol_eig.extend((zz[1],zz[3],zz[5],zz[7]))

    return sol_real,sol_eig,t


```

```python
P=0.1 ############### TUNING parameter
g_y=1e-1
N0=(P*del_omega)**3/(8*g_y**2)

title='$g_y={0}\Omega  \: ; \: N_0={1} \: ; \: P_c={2}$'.format(g_y,np.round(N0),P)
label_list=["$\psi_1$","$\psi_1*$","$\psi_3$","$\psi_3*$","$A_0^+$","$A_0^-$","$q_z$","$p_z$"]
tf=5e0
N_step=10000
qy_0=1e-2
param_list=[g_z1,g_y,del_omega,N0]   ##g_z0,g_z1,g_y,ohm, omega_0, omega_1, del_omega, N_eq
# fig, ax = plt.subplots(5, 2,figsize=(18,26))
# fig.suptitle(title, fontsize=20)
om_list=2*np.pi*np.arange(N_step)/tf
sf,ef=1,400
z0=np.array([1,1,-1,-1,0,0,0,0,0])*qy_0+0j ## prefactor of delta*g_y/(2*D) multiplied in other function
z0[8]=1
yy1,zz1,t1=psi_evol(z0,tf,param_list,N_step)
z0=np.array([1,1,-1,-1,0,0,0,0])*qy_0+0j ## prefactor of delta*g_y/(2*D) multiplied in other function
yy2,zz2,t2=psi_evol_app(z0,tf,param_list,N_step)
index=1
print(np.shape(yy2[index]),np.shape(t2))
plt.plot(t1,yy1[index],label="full")
plt.plot(t2,yy2[index],label="approx")
plt.ylabel(label_list[index],fontsize=14)
plt.xlabel('$\Omega t$',fontsize=14)
plt.legend()
plt.show()
yf1=np.log(np.abs(np.fft.rfft(yy1[index]))/N_step)## Fourier transform
yf2=np.log(np.abs(np.fft.rfft(yy2[index]))/N_step)## Fourier transform

plt.plot(om_list[sf:ef],yf1[sf:ef],label="full")
plt.plot(om_list[sf:ef],yf2[sf:ef],label="approx")
plt.ylabel('$\mathcal{F}[$'+label_list[index]+"]",fontsize=14)
plt.xlabel('$\omega$',fontsize=14)
plt.legend()
plt.show()
```
